## Tesseract

```python
def read_png(file_path):
    if not os.path.exists(file_path):
        print("File path does not exist")
        return
    try:
        img=Image.open(file_path)
        text=pytesseract.image_to_string(img,lang='eng')
        print(text)
    except Exception as e:
        print(e)
```
