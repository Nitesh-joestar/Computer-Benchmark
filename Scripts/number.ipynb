```python
import numpy as np
import cv2
from PIL import ImageGrab
import pytesseract
import time
import keyboard
import pyautogui as p
custom_config = r'--psm 7 -c tessedit_char_whitelist=0123456789'
```

```python
img_box=(0,202,1345,492)
```

```python
def capture_to_memory(img_box):
    #time.sleep(1)
    cap=ImageGrab.grab(bbox=img_box)
    img_np=np.array(cap)
    frame=cv2.cvtColor(img_np,cv2.COLOR_RGB2GRAY)#below converts to black and white
    (_, blackAndWhiteImage) = cv2.threshold(frame, 127, 255, cv2.THRESH_BINARY_INV)
    padded = cv2.copyMakeBorder(blackAndWhiteImage, 10, 10, 50, 10, cv2.BORDER_CONSTANT, value=[255, 255, 255])
    cv2.imshow("text",padded)
    cv2.waitKey(1000)
    cv2.destroyAllWindows()
    text=pytesseract.image_to_string(padded,config=custom_config)
    return text.replace('\n','')
```

```python
capture_to_memory(img_box=img_box)
```

```python
p.position()
```

```python
def write_number(num):
    p.moveTo(652,397)
    p.click()
    p.write(num)
    time.sleep(1)
    p.moveTo(679,496)
    p.click()
    time.sleep(3)
    p.moveTo(675,549)
    p.click()
```

```python
p.moveTo(672,548)
p.click()
sleeper=3
while keyboard.is_pressed('q') is False:
    if(p.pixel(539,586)==(115, 84, 0)):
        break
    text=capture_to_memory(img_box=img_box)
    time.sleep(sleeper)
    sleeper+=1
    write_number(text)
```
