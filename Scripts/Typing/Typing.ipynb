```python
import numpy as np
import cv2
from PIL import ImageGrab
import pytesseract
import time
import pyautogui as p
```

```python
img_box=(197,393,1148,519)
```

```python
def capture_to_memory(img_box):
    time.sleep(2)
    cap=ImageGrab.grab(bbox=img_box)
    img_np=np.array(cap)
    frame=cv2.cvtColor(img_np,cv2.COLOR_RGB2GRAY)
    text=pytesseract.image_to_string(frame)
    return text
```

```python
text=capture_to_memory(img_box=img_box)
clean_text=text.replace('\n',' ')
print(clean_text)
```

```python
time.sleep(2)
p.position()
```

```python
p.moveTo(229,409,duration=2)
p.click()
p.write(clean_text)
```
