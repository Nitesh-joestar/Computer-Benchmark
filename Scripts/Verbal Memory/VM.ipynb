```python
import numpy as np
import cv2
from PIL import ImageGrab
import pytesseract
import time
import keyboard
import pyautogui as p
custom_config = r'--psm 1 -c tessedit_char_whitelist=0123456789'
```

```python
p.position()
```

```python
image_box=(393,337,956,444)
```

```python
def capture_to_memory(img_box):
    #time.sleep(1)
    cap=ImageGrab.grab(bbox=img_box)
    img_np=np.array(cap)
    frame=cv2.cvtColor(img_np,cv2.COLOR_RGB2GRAY)#below converts to black and white
    (_, blackAndWhiteImage) = cv2.threshold(frame, 100, 255, cv2.THRESH_BINARY)
    padded = cv2.copyMakeBorder(blackAndWhiteImage, 10, 10, 50, 10, cv2.BORDER_CONSTANT, value=[0,0,0])
    cv2.imshow("tdext",padded)
    cv2.waitKey(1)
    cv2.destroyAllWindows()
    text=pytesseract.image_to_string(padded)
    return text.replace('\n','')
```

```python
capture_to_memory(img_box=image_box)
```

```python
words=[]
def check(text):
    if text in words:
        return True
    else:
        return False
```

```python
p.click(760,482)
```

```python
words.clear()
p.moveTo(643,572)
p.click()
while keyboard.is_pressed('q') is False:
    time.sleep(1)
    text=capture_to_memory(img_box=image_box)
    if text in words:
        p.click(589,485)  
    else:
        p.click(760,482)
    words.append(text)
    print(words)
```
